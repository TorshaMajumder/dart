```python
import os
import umap
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.palettes import plasma
from bokeh.plotting import figure
from bokeh.transform import transform
from bokeh.io import output_file, show, save
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper, WheelZoomTool, ColorBar
```

## Initial Training with Astronomaly

```python
ml_score_1 = pd.read_csv(f"C:/Users/torsh/git/TESS/SVM/astronomaly/score_/ml_scores_1.csv")
ml_score_1.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
ml_score_1.head()
```

```python
umap_data = pd.read_parquet(f"C:/Users/torsh/git/TESS/SVM/astronomaly/score_/UMAP_Plot_output.parquet", engine="pyarrow")
umap_data.rename(columns = {'0':'feature_1', '1':'feature_2'}, inplace = True)
umap_data = umap_data.reset_index()
umap_data.head()
```

```python
new_data_df = pd.merge(ml_score_1,umap_data, on="index", how="inner") 
new_data_df.head(10)
```

```python
source = ColumnDataSource(data=dict(x=new_data_df["feature_1"], y=new_data_df["feature_2"], desc=new_data_df["index"], score=new_data_df["score"]))
hover = HoverTool(tooltips=[
        ('Label:', '@desc'),
        ('Score:', '@score'),
        ])
mapper = LinearColorMapper(palette=plasma(256), low=min(new_data_df["score"]), high=max(new_data_df["score"]))
p = figure(plot_width=1000, plot_height=800, tools=[hover, WheelZoomTool()], title=f"Cluster_-1 using Astronomaly (retraining with ~200 samples)", toolbar_location = "below")
p.circle("x", "y",size=10, source=source, line_color=None, fill_color={"field":"score","transform":mapper})
bar = ColorBar(color_mapper=mapper, location=(0,0))
p.add_layout(bar, "left")
output_file(f"C:/Users/torsh/git/TESS/SVM/astronomaly/cluster_-1_train1.html")
save(p)
#show(p)
```
